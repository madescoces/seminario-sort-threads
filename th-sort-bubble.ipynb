```python
logs = []
```

```python
sort = BubbleSort()
th2_bubble_sort = sort.run(data)
logs.append(sort.log())

print(th2_bubble_sort[:5], end=' ')
```

```python
sort = BubbleSort(threads = 2)
th4_bubble_sort = sort.run(data)
logs.append(sort.log())

print(th4_bubble_sort[:5], end=' ')
```

```python
sort = BubbleSort(threads = 12)
th12_bubble_sort = sort.run(data)
logs.append(sort.log())

print(th12_bubble_sort[:5], end=' ')
```

```python
sort = MergeSort()
th2_merge_sort = sort.run(data)
logs.append(sort.log())

print(th2_merge_sort[:5], end=' ')
```

```python
bubble_sort_data = [entry for entry in logs if entry['name'] == 'BubbleSort']

# Ordenar los datos seg칰n el n칰mero de cores
bubble_sort_data.sort(key=lambda x: x['cores'])

# Extraer las coordenadas x e y para cada algoritmo
merge_x = [entry['cores'] for entry in bubble_sort_data]
merge_y = [entry['time'] for entry in bubble_sort_data]

# Graficar los datos
plt.plot(merge_x, merge_y, marker='o', label='BubbleSort', linestyle='-', color='blue')

# Agregar etiquetas y t칤tulo
plt.xlabel('N칰mero de Cores')
plt.ylabel('Tiempo (segundos)')
plt.title('Tiempo de Ejecuci칩n entre BubleSort - 20000 registros')

# Mostrar leyenda
plt.legend()

# Agregar 칰ltimo log como etiqueta
last_log = logs[-1]
plt.text(6, 12, f"{last_log['name']} 1 core - {last_log['time']:.2f} segundos", ha='left', va='top', color='red', backgroundcolor='white', bbox=dict(facecolor='white', edgecolor='white', boxstyle='round,pad=0.5'))

# Mostrar la gr치fica
plt.show()
```
